```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')
plt.subplot(122)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
```

pie(x, explode=None, labels=None, colors=None, autopct-none, pctdistance=0. 6,
shadow=False, labeldistance=1. 1, startangle-none, radius=None)
参数：x（每一块）的比例，如果sum(x)>1会使用sum(x)归化
labels（每一块）饼图外侧显示的说明文字
explode（每一块）离开中心距离
strangle i起始绘制角度，默认图是从x轴正方向逆时针画起如设定=90则从y轴正方向画起
shadow是否阴影
labeldistance labels绘制位置，相对于半径的比例，如<1则绘制在饼图内侧
autopct控制饼图内百分比设置，可以使用 iformat字符串或者 format function
'%1.1f'指小数点前后位数没有用空格补齐）
pctdistance类似于 labeldistance,指定 autopct的位置刻度
radius控制饼图半径
返回值：如果没有设置 autopct返回( patches, texts)
如果设置 autopct,返回( patches, texts, autotext)

```python
data1=pd.cut(df['EventOrigin_x'],[0,20,40,60,80,100]).value_counts()
data1
```

```python
y=data1.values
#y=y/sum(y)
plt.figure(figsize=(5,5))
plt.title('测试3',fontsize=18)
plt.pie(y,labels=data1.index,autopct='%.lf %%',colors='byrcg',startangle=180,radius=1,shadow=True)  #第1 2%为对格式的设置
plt.legend()
plt.show()
```

```python
patches,l_text,p_text=plt.pie(y,labels=data1.index,autopct='%.lf %%',colors='byrcg',startangle=180,radius=1,shadow=True)  #第1 2%为对格式的设置
for i in p_text:
    i.set_size(15)
    i.set_color('w')
for i in l_text:
    i.set_size(15)
    i.set_color('w')
plt.legend()
plt.show()
```

bins:直方图的柱数，可选项，默认为10
normed:是否将得到的直方图向量归一化。默认为0
facecolor.直方图颐色
edgecolor直方图边框颜色
alpha:透明度
histotype:直方图类型，"bar, "barstacked,'step,' stepfilled
返回值
n:直方图向量，是否归化由参数 normed设定
bins:返回各个bin的区间范围
patches:返回每个bin里面包含的数据，是个list
