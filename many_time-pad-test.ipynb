```python
import numpy as np 


def xor_char(ch1, ch2):
    res = ord(ch1) ^ ord(ch2)
    print(hex(res))
    

def extract_ciphertext(ciphertext: str):
    res_list = []
    for i in range(0, len(ciphertext), 2):
        single_hex_val = ciphertext[i:i+2]
        res_list.append(int(single_hex_val, 16))
    return res_list 


def is_alphabet_or_space(x):
    if ord('a') <= x and x <= ord('z'): 
        return True
    if ord('A') <= x and x <= ord('Z'): 
        return True
    if x == ord(' '):
        return True
    return False


def convert_to_str(val_list):
    original_str = ''
    for val in val_list:
        original_str += chr(val)
    return original_str


def is_originally_alphabet_or_space(x):
    if is_alphabet_or_space(x ^ ord(' ')):
        return True
    return False


def extract_ciphertext(ciphertext: str):
    res_list = []
    for i in range(0, len(ciphertext), 2):
        single_hex_val = ciphertext[i:i+2]
        res_list.append(int(single_hex_val, 16))
    return res_list


def attack(ciphertext_list):
    all_ciphertext_val_list = [extract_ciphertext(ciphertext) for ciphertext in ciphertext_list]
    key_length = min(len(ciphertext_val_list) for ciphertext_val_list in all_ciphertext_val_list)
    
    plain_text_mat = np.zeros((len(all_ciphertext_val_list), key_length), dtype=int)
    
    for idx, ciphertext_val_list in enumerate(all_ciphertext_val_list):
        is_space = []
        for j in range(key_length):
            score = 0
            for target_ciphertext_val_list in all_ciphertext_val_list:
            
                xor_val = ciphertext_val_list[j] ^ target_ciphertext_val_list[j]
                if is_originally_alphabet_or_space(xor_val):
                    score += 1
            if score >= 8:
                is_space.append(True)
            else:
                is_space.append(False)
        for space_idx, if_is_space in enumerate(is_space):
            if if_is_space:
                # 利用 第idx个明文的第space_idx个字符是空格 恢复所有明文的这一列
                for target_idx, target_ciphertext_val_list in enumerate(all_ciphertext_val_list):
                    plain_text_mat[target_idx, space_idx] = ciphertext_val_list[space_idx] ^ target_ciphertext_val_list[space_idx] ^ ord(' ')
    
    return plain_text_mat
```
